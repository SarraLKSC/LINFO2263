```python
#NOW LET'S COMPUTE idf_j
N=len(words)
for key,value in idf_j.items():
  idf_j[key]=math.log10(N/value)
```

```python
#NOW WE MULTIPLY idf_j x tf_ij
i=0
w_ij={}
for key,value in tf_ij.items():
  context = key[1]
  w_ij[key]= value* idf_j[context]
```

```python
def extract_semantic_vector(target):
  semantic_vec=np.zeros(len(vocabulary))
  for key,value in w_ij.items():
    if (key[0]!= target):
      continue
    idx=vocabulary.index(key[1])
    semantic_vec[idx]=value
  return semantic_vec
```

```python
extract_semantic_vector('glimpse')
```

```python
def cos_similarity(w,v):
  w_vec=extract_semantic_vector(w)
  v_vec=extract_semantic_vector(v)
  top=0
  for i in range(len(vocabulary)):
    top+= w_vec[i] * v_vec[i]
  result= top / (np.linalg.norm(w_vec,2) * np.linalg.norm(v_vec,2))
  return result
```

```python
def top_five(target):
  all=[cos_similarity(target,v) for v in vocabulary]
  all =np.array(all)
  idx_best=all.argsort()[-6:][::-1]
  return ([vocabulary[i] for i in idx_best[1:]])
```
